##### Nasjonalt register for opplæringskontor: Alle organisasjoner
NOR består av  organisasjoner. Vi skal hente ut detaljert informasjon om samtlige av dem. 

```python
import requests
import pandas as pd
import json

# Locale module
from nxr_classes import Organisations
```

```python
nor_organisations = Organisations('nor')

nor_organisations.get_overview()

nor_organisations.overview.sample(10)
```

```python
for org in nor_organisations.overview['Orgnr']:
    nor_organisations.get_org(org)
    nor_organisations.flatten_json(org)


nor_organisations.gen_df()

cols = nor_organisations.df.columns.tolist()
first_cols = list()
end_cols = list()

for col in cols:
    if "Naeringskoder" in col:
        end_cols.append(col)
    else:
        first_cols.append(col)

new_cols = first_cols + end_cols

nor_organisations.df = nor_organisations.df[new_cols]
```
