```python
import requests
import pandas as pd
import json

# Locale module
from nxr_classes import Organisations
```

```python
##### Nasjonalt barnehageregister: Alle organisasjoner
NBR består av 14 924 organisasjoner. Vi skal hente ut detaljert informasjon om samtlige av dem. 
```

```python
nbr_organisations = Organisations('nbr')

nbr_organisations.get_overview()

nbr_organisations.overview.sample(10)
```

```python
for org in nbr_organisations.overview['Orgnr']:
    nbr_organisations.get_org(org)
    nbr_organisations.flatten_json(org)


nbr_organisations.gen_df()

cols = nbr_organisations.df.columns.tolist()
first_cols = list()
end_cols = list()

for col in cols:
    if "Barnehagekategorier" in col or  "Naeringskoder" in col:
        end_cols.append(col)
    else:
        first_cols.append(col)

new_cols = first_cols + end_cols

nbr_organisations.df = nbr_organisations.df[new_cols]

```

```python
nbr_organisations.df.to_excel(nbr_organisations.register +" - alle organisasjoner.xlsx")
```
